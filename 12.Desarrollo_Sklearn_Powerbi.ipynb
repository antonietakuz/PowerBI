```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dataset=pd.read_csv('D:/Mis cosas/PowerBI/Nueva carpeta (2)/HR_comma_sep2023.csv', sep=',', engine='python')
```

```python
dataset
```

```python
le=LabelEncoder()
```

```python
from sklearn import preprocessing
```

```python
dataset['salary']=le.fit_transform(dataset['salary'])
```

```python
dataset['Department']=le.fit_transform(dataset['Department'])
```

```python
y=dataset['left']
features=['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours', 'time_spend_company', 'Work_accident', 'promotion_last_5years','Department','salary']
```

```python
X=dataset[features]
```

```python
s=StandardScaler()
X=s.fit_transform(X)
```

```python
X_train,X_test, y_train, y_test=train_test_split(X, y)
```

```python
log=LogisticRegression()
log.fit(X_train, y_train)
y_pred=log.predict(X)
y_prob=log.predict_proba(X)[:,1]
```
